```python
# Import dataset
d = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Remove customer IDs from the data set
d = d.iloc[:,1:]

# Remove all seniors from dataset
d = d[d['SeniorCitizen'] == 0]
d = d.drop(columns='SeniorCitizen')
```

```python
# Churn subset excluding senior citizens
ax=(d['Churn'].value_counts()*100.0 /len(d)).plot.pie(autopct='%.1f%%', labels=['No', 'Yes'], figsize =(5,5), fontsize = 12 )                                                                           
ax.set_title('Churn subset excluding senior citizens')
```
